```python
#Scraping the website(url: https://www.4icu.org/in/) and store the details of collages in a CSV file


import requests
import bs4
import csv

res=requests.get('https://www.4icu.org/in/')
soup=bs4.BeautifulSoup(res.text,'lxml')

s1=soup.select('th')
s2=soup.select('td')

s2=s2[1:2635]

s3=s1 + s2


list=[]
j=0

for i in s3:
    if(j%3==0):
        rank=i.text
    elif(j%3==1):
        university=i.text
    else:
        town=i.text
    j+=1
    if(j%3==0):
        details=[rank,university,town]
        list.append(details)

with open('2018_Indian_University_Ranking.csv','w',newline="") as file:
    write=csv.writer(file)
    for row in list:
        write.writerow(row)
```
