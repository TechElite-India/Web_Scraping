```python
import requests
import bs4
import csv

list=[]
lis_h=['Course_Name','Link']
list.append(lis_h)

for j in range(29):
    res1=requests.get('https://www.studyinnewzealand.govt.nz/study-options/course/provider-results?institutionid=142314&pageno='+str(j)) 
    soup1=bs4.BeautifulSoup(res1.text,'lxml')
    s1=soup1.select('h3')
    for i in range(len(s1)):
        if(s1[i].get('class')==['crs_tit', 'univ_tit']):
            list.append([s1[i].a.text,s1[i].a.get('href')])
            
list
```

```python
len(list)
```

```python
with open('Auckland_University_of_Technology.csv','w',newline="") as file:
    write=csv.writer(file)
    for row in list:
        write.writerow(row)
```
