```python
import requests
import bs4
import csv

list=[]
lis_h=['Course_Name','Link']
list.append(lis_h)

res1=requests.get('http://www.lincoln.ac.nz/Study/Courses/') 
soup1=bs4.BeautifulSoup(res1.text,'lxml')

s1=soup1.select('a')
for i in s1[3:]:
    list.append([str(i.text)[11:],i.get('href')])
    
list
```

```python
len(list)
```

```python
with open('Lincoln_University .csv','w',encoding='latin-1',newline="") as file:
    write=csv.writer(file)
    for row in list:
        write.writerow(row)
```
