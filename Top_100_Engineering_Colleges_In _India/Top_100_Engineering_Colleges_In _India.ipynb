```python
#Scraping the website(url: https://www.outlookindia.com/magazine/story/top-100-engineering-colleges-in-india/231883) and store the ranking details of engineering collages in a CSV file

import requests
import bs4
import csv

data=requests.get('https://www.outlookindia.com/magazine/story/top-100-engineering-colleges-in-india/231883')
soup1=bs4.BeautifulSoup(data.text,'lxml')

data1=soup1.select('p')
data1=data1[4:1004]

data_title_list=["Rank","Name of Institute","City","Government(G)/Private(P)","Intellectual_Capital(30)","Infrastructure_and_Facilities(20)","Pedagogic_Systems(15)","Industry_Interface(15)","Placements(20)","Total(100)"]
data_list=[]
data_list.append(data_title_list)
j=0

for i in data1:
    if(j%10==0):
        rank=i.text
    elif(j%10==1):
        name_of_inst=i.text
    elif(j%10==2):
        city=i.text
    elif(j%10==3):
        G_or_P=i.text
    elif(j%10==4):
        IC=i.text
    elif(j%10==5):
        I_and_F=i.text
    elif(j%10==6):
        PS=i.text
    elif(j%10==7):
        II=i.text
    elif(j%10==8):
        P=i.text
    else:
        total=i.text
    j+=1
    if(j%10==0):
        combined_list=[rank,name_of_inst,city,G_or_P,IC,I_and_F,PS,II,P,total]
        data_list.append(combined_list)
    
with open('Top_100_Engineering_Colleges_In _India.csv','w',newline="") as file:
    write=csv.writer(file)
    for row in data_list:
        write.writerow(row) 
```
