```python
import requests
import bs4
import csv

list=[]
lis_h=['Course_Name','Course_Level','Course_Type','Cource_Description','University_Name','Link']
list.append(lis_h)

res2=requests.get('https://www.handbook.unsw.edu.au/') 
soup2=bs4.BeautifulSoup(res2.text,'lxml')
s2=soup2.select('a')
for j in s2[29:167]:
    res1=requests.get('https://www.handbook.unsw.edu.au'+j.get('href')) 
    soup1=bs4.BeautifulSoup(res1.text,'lxml')
    s1=soup1.select('#subjectUndergraduate')
    if(len(s1)!=0):
        s11=s1[0].select('a')
        for i in range(len(s11)-1):
            if(len(s11[i].select('div'))!=2):
                list.append([s11[i].select('div')[3].text,'Undergraduate',str(j.h4.text)[6:],'','University of New South Wales','https://www.handbook.unsw.edu.au'+s11[i].get('href')[:-1]])
            else:
                list.append([str(s11[i].select('div')[1].text)[1:-1],'Undergraduate',str(j.h4.text)[6:],'','University of New South Wales','https://www.handbook.unsw.edu.au'+s11[i].get('href')[:-1]])

    s1=soup1.select('#subjectPostgraduate')
    if(len(s1)!=0):
        s11=s1[0].select('a')
        for i in range(len(s11)-1):
            if(len(s11[i].select('div'))!=2):
                list.append([s11[i].select('div')[3].text,'Postgraduate',str(j.h4.text)[6:],'','University of New South Wales','https://www.handbook.unsw.edu.au'+s11[i].get('href')[:-1]])
            else:
                list.append([str(s11[i].select('div')[1].text)[1:-1],'Postgraduate',str(j.h4.text)[6:],'','University of New South Wales','https://www.handbook.unsw.edu.au'+s11[i].get('href')[:-1]])
list
```

```python
len(list)
```

```python
with open('University_of_New_South_Wales.csv','w',newline="") as file:
    write=csv.writer(file)
    for row in list:
        write.writerow(row)
```
