```python
import requests
import bs4
import csv

list=[]
lis_h=['Course_Name','Course_Level','Course_Type','Cource_Description','University_Name','Link']
list.append(lis_h)

res=requests.get('http://www.deakin.edu.au/courses/find-a-course/choose-your-future')
soup=bs4.BeautifulSoup(res.text,'lxml')

s=soup.select('a')
for i in s[305:748]:
    if(i.text!='More Info'):
        res1=requests.get(i.get('href'))
        soup1=bs4.BeautifulSoup(res1.text,'lxml')
        
        st2=''
        s2=soup1.select('.module__banner-description')
        for a in str(s2[0].text)[1:]:
            if(a=='\n'):
                break
            st2+=a
        
        list.append([i.text,'Undergraduate','',st2,'Deakin University',i.get('href')])
        
list
```

```python
with open('Deakin_University.csv','w',newline="") as file:
    write=csv.writer(file)
    for row in list:
        write.writerow(row)
```

```python
list1=[]
res=requests.get('http://www.deakin.edu.au/courses/find-a-course/postgraduate')
soup=bs4.BeautifulSoup(res.text,'lxml')

s=soup.select('td')
s1=soup.select('a')
for i in range(450):
    
    res1=requests.get(s1[i+302].get('href'))
    soup1=bs4.BeautifulSoup(res1.text,'lxml')
    
    st2=''
    s3=soup1.select('.module__banner-description')
    for a in str(s3[0].text)[1:]:
        if(a=='\n'):
            break
        st2+=a
    
    list1.append([s[i*3].text,'Postgraduate','',st2,'Deakin University',s1[i+302].get('href')])
list1
```
