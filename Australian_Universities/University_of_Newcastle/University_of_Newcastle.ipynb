```python
import requests
import bs4
import csv

list=[]
lis_h=['Course_Name','Course_Level','Course_Type','Cource_Description','University_Name','Link']
list.append(lis_h)
```

```python
res=requests.get('https://www.newcastle.edu.au/degrees')
soup=bs4.BeautifulSoup(res.text,'lxml')
```

```python
s=soup.select('.title ')
for i in range(len(s)):
    print(s[i].text)
    print('https://www.newcastle.edu.au/'+s[i].a.get('href'))
```

```python
len(s)
```

```python
s=soup.select('.title ')
for i in range(len(s)-1):
    list.append([s[i].text,'','','','University of Newcastle','https://www.newcastle.edu.au/'+s[i].a.get('href')])
    
list
```

```python
len(list)
```

```python
with open('University_of_Newcastle.csv','w',newline="") as file:
    write=csv.writer(file)
    for row in list:
        write.writerow(row)
```
