```python
import requests
import bs4
import csv

res1=requests.get('https://www.monash.edu/study/courses')
soup=bs4.BeautifulSoup(res1.text,'lxml')

s=soup.select('.area-blocks__link')

list=[]
lis_h=['Course_Name','Course_Level','Course_Type','Cource_Description','University_Name','Link']
list.append(lis_h)

for j in s:
    res1=requests.get(j.get('href'))
    soup=bs4.BeautifulSoup(res1.text,'lxml')
    
    s1=soup.select('.box-featured__heading-link')
    s2=soup.select('.box-featured__level')
    s3=soup.select('.box-featured__description')
    
    
    for i in range(len(s1)-1):
        lis=[s1[i].text[7:(len(s1[i].text)-6)],s2[i].text,j.text,s3[i].text[1:(len(s3[i].text)-3)],'Monash University',s1[i].get('href')]
        list.append(lis)

list
```

```python
len(list)
```

```python
with open('Monash_University.csv','w',newline="") as file:
    write=csv.writer(file)
    for row in list:
        write.writerow(row)
```
