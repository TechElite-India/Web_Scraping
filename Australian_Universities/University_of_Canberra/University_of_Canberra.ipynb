```python
import requests
import bs4
import csv

list=[]
lis_h=['Course_Name','Course_Level','Course_Type','Cource_Description','University_Name','Link']
list.append(lis_h)

res=requests.get('https://search.canberra.edu.au/s/search.html?collection=courses&form=course-search&profile=_default&query=!padre&course-search-widget__submit=&meta_C_and=COURSE&sort=metaH&start_rank=1')
soup=bs4.BeautifulSoup(res.text,'lxml')
s=soup.select('tbody')
s11=s[0].select('a')
for i in range(len(s11)):
    list.append([str(s11[i].text)[19:-17],'','','','University of Canberra',s11[i].get('title')])
    
for j in range(1,75):
    res1=requests.get('https://search.canberra.edu.au/s/search.html?collection=courses&form=course-search&profile=_default&query=!padre&course-search-widget__submit=&meta_C_and=COURSE&sort=metaH&start_rank='+str(j)+str(1))
    soup1=bs4.BeautifulSoup(res1.text,'lxml')
    s1=soup1.select('tbody')
    s22=s1[0].select('a')
    for i in range(len(s22)):
        list.append([str(s22[i].text)[19:-17],'','','','University of Canberra',s22[i].get('title')])
        
list
```

```python
with open('University_of_Canberra.csv','w',newline="") as file:
    write=csv.writer(file)
    for row in list:
        write.writerow(row)
```
