```python
import requests
import bs4
import csv

list=[]
lis_h=['Course_Name','Course_Level','Course_Type','Cource_Description','University_Name','Link']
list.append(lis_h)

res1=requests.get('http://www.anu.edu.au/study/study-options')
soup1=bs4.BeautifulSoup(res1.text,'lxml')
s1=soup1.select('.extraspace')
for i in range(len(s1)):
    for j in s1[i].select('a'):
        res2=requests.get(j.get('href'))
        soup2=bs4.BeautifulSoup(res2.text,'lxml')
        s2=soup2.select('.field-items')
        s22=s2[0].select('ul')
        if(len(s22)!=0):
            for k in range(len(s22)):
                for l in s22[k].select('a'):
                    list.append([l.text,'',str(j.text)[:-2],'','Australian National University',l.get('href')])
                    
list
```

```python
len(list)
```

```python
with open('Australian_National_University.csv','w',newline="") as file:
    write=csv.writer(file)
    for row in list:
        write.writerow(row)
```
