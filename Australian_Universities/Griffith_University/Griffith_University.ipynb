```python
import requests
import bs4
import csv

res=requests.get('https://www.griffith.edu.au/study?location=intl')
soup=bs4.BeautifulSoup(res.text,'lxml')

list=[]
lis_h=['Course_Name','Course_Level','Course_Type','Cource_Description','University_Name','Link']
list.append(lis_h)

s=soup.select('a')
for i in s[315:325]:
    res1=requests.get(i.get('href'))
    soup1=bs4.BeautifulSoup(res1.text,'lxml')
    
    s1=soup1.select('li')
    for j in s1[172:-61]:
        if(j.text!='Architecture, construction and planning' and j.text!='I have read and accept the conditions outlined in the privacy plan. (required)'):
            res2=requests.get(j.a.get('href'))
            soup2=bs4.BeautifulSoup(res2.text,'lxml')
            s2=soup2.select('table')
            for a in range(len(s2)):
                for b in s2[a].select('a'):
                    list.append([b.text,'',j.text,'','Griffith University',b.get('href')])
```

```python
with open('Griffith_University.csv','w',newline="") as file:
    write=csv.writer(file)
    for row in list:
        write.writerow(row)
```
