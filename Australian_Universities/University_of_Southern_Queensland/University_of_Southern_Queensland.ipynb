```python
import requests
import bs4
import csv

list=[]
lis_h=['Course_Name','Course_Level','Course_Type','Cource_Description','University_Name','Link']
list.append(lis_h)

res1=requests.get('https://www.usq.edu.au/study')
soup1=bs4.BeautifulSoup(res1.text,'lxml')
s1=soup1.select('.degree-navigation__bottom')
s11=s1[0].select('a')
for k in s11[:-3]:
    res=requests.get('https://www.usq.edu.au'+k.get('href'))
    soup=bs4.BeautifulSoup(res.text,'lxml')
    s=soup.select('.panel-group')
    for j in range(len(s)):
        for i in s[j].select('a'):
            if(str(i.get('href'))[0]!='#'and i.text!='Find out more...'):
                list.append([i.text,'',str(k.text)[1:-2],'','University of Southern Queensland','https://www.usq.edu.au'+i.get('href')])

list
```

```python
len(list)
```

```python
list1=[]
res2=requests.get('https://www.usq.edu.au/study/degrees/professional-development')
soup2=bs4.BeautifulSoup(res2.text,'lxml')
s2=soup2.select('ul')
for l in s2[5].select('a')[1:]:
    res3=requests.get(l.get('href'))
    soup3=bs4.BeautifulSoup(res3.text,'lxml')
    s3=soup3.select('table')
    for m in s3[0].select('a'):
        list1.append([str(m.text)[8:],'',l.text,'','University of Southern Queensland',m.get('href')])
        
list1
```

```python
len(list1)
```

```python
list2=list+list1
len(list2)
```

```python
with open('University_of_Southern_Queensland.csv','w',newline="") as file:
    write=csv.writer(file)
    for row in list2:
        write.writerow(row)
```
