```python
import requests
import bs4
import csv

list=[]
lis_h=['Course_Name','Course_Level','Course_Type','Cource_Description','University_Name','Link']
list.append(lis_h)

lk=['Undergraduate(Local)','Postgraduate(Local)','Undergraduate(International)','Postgraduate(International)']

res=requests.get('https://www.mq.edu.au/study/find-a-course')
soup=bs4.BeautifulSoup(res.text,'lxml')
s=soup.select('.list-4-12')
for i in s[0].select('li'):
    res1=requests.get(i.a.get('href'))
    soup1=bs4.BeautifulSoup(res1.text,'lxml')
    s1=soup1.select('.subject-listing__item')
    for j in range(len(s1)):
        res2=requests.get(s1[j].a.get('href'))
        soup2=bs4.BeautifulSoup(res2.text,'lxml')
        s2=soup2.select('.accordion')
        if(len(s2)>=4):
            for l in range(4):
                for k in s2[l].select('a'):
                    list.append([k.text,lk[l],s1[j].a.text,'','Macquarie University',k.get('href')])

list
```

```python
len(list)
```

```python
with open('Macquarie_University.csv','w',encoding="utf-8",newline="") as file:
    write=csv.writer(file)
    for row in list:
        write.writerow(row)
```
