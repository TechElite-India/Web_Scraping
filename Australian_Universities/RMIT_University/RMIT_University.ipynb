```python
import requests
import bs4
import csv

list=[]
lis_h=['Course_Name','Course_Level','Course_Type','Cource_Description','University_Name','Link']
list.append(lis_h)

res=requests.get('https://www.rmit.edu.au/study-with-us')
soup=bs4.BeautifulSoup(res.text,'lxml')

s=soup.select('.cta')
for i in s[:16]:
    res1=requests.get('https://www.rmit.edu.au'+i.get('href'))
    soup1=bs4.BeautifulSoup(res1.text,'lxml')
    s1=soup1.select('.masthead-child-links-list')
    s11=s1[0].select('a')
    for j in range(len(s11)-1):
        res2=requests.get('https://www.rmit.edu.au'+s11[j].get('href'))
        soup2=bs4.BeautifulSoup(res2.text,'lxml')
        s2=soup2.select('.m-program-list-title')
        for k in range(len(s2)):
            list.append([str(s2[0].a.text)[7:-5],s2[k].div.text,i.text,'','RMIT University','https://www.rmit.edu.au'+s2[k].a.get('href')])
            
list            
```

```python
len(list)
```

```python
with open('RMIT_University.csv','w',newline="") as file:
    write=csv.writer(file)
    for row in list:
        write.writerow(row)
```
