```python
import requests
import bs4
import csv

list=[]
lis_h=['Course_Name','Course_Level','Course_Type','Cource_Description','University_Name','Link']
list.append(lis_h)
```

```python
res1=requests.get('http://www.ecu.edu.au/degrees-elements/courses-list') 
soup1=bs4.BeautifulSoup(res1.text,'lxml')
```

```python
s1=soup1.select('a')
for i in s1[152:399]:
    list.append([i.text,'','','','Edith Cowan University',i.get('href')])
list
```

```python
len(list)
```

```python
with open('Edith_Cowan_University.csv','w',newline="") as file:
    write=csv.writer(file)
    for row in list:
        write.writerow(row)
```
