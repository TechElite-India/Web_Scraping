```python
import requests
import bs4
import csv

list=[]
lis_h=['Course_Name','Course_Level','Course_Type','Cource_Description','University_Name','Link']
list.append(lis_h)

res=requests.get('https://www.scu.edu.au')
soup=bs4.BeautifulSoup(res.text,'lxml')

s=soup.select('#decoration')
for i in range(len(s)):
    res1=requests.get('https://www.scu.edu.au'+s[i].get('href'))
    soup1=bs4.BeautifulSoup(res1.text,'lxml')
    s1=soup1.select('.pagination')
    if(len(s1)!=0):
        s11=s1[0].select('a')
        for j in range(len(s11)-2):
                res2=requests.get('https://www.scu.edu.au'+s11[j].get('href'))
                soup2=bs4.BeautifulSoup(res2.text,'lxml')
                s2=soup2.select('tbody')
                s22=s2[0].select('a')
                for j in range(len(s22)):
                    if(str(s22[j].text)[8:-10]!=''):
                        list.append([str(s22[j].text)[8:-10],'',str(s[i].text)[5:-3],'','Southern Cross University','https://www.scu.edu.au'+s22[j].get('href')])
    else:
        res2=requests.get('https://www.scu.edu.au'+s[i].get('href'))
        soup2=bs4.BeautifulSoup(res2.text,'lxml')
        s2=soup2.select('tbody')
        s22=s2[0].select('a')
        for j in range(len(s22)):
            if(str(s22[j].text)[8:-10]!=''):
                list.append([str(s22[j].text)[8:-10],'',str(s[i].text)[5:-3],'','Southern Cross University','https://www.scu.edu.au'+s22[j].get('href')])
        
list
```

```python
with open('Southern_Cross_University.csv','w',newline="") as file:
    write=csv.writer(file)
    for row in list:
        write.writerow(row)
```
