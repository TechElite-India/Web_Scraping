```python
import requests
import bs4
import csv

list=[]
lis_h=['Course_Name','Course_Level','Course_Type','Cource_Description','University_Name','Link']
list.append(lis_h)

res=requests.get('https://www.flinders.edu.au/')
soup=bs4.BeautifulSoup(res.text,'lxml')


s=soup.select('a')
for i in s[4:23]:
    
    res1=requests.get('https://www.flinders.edu.au'+str(i.get('href')))
    soup1=bs4.BeautifulSoup(res1.text,'lxml')
    s1=soup1.select('.item_heading')
    s2=soup1.select('.item_content')
    
    for j in range(len(s1)):
        for b in s2[j].ul.select('li'):
            list.append([str(b.a.text)[2:-1],str(s1[j].text)[6:-1],i.text,'','Flinders University',b.a.get('href')])
    
list
```

```python
len(list)
```

```python
with open('Flinders_University.csv','w',newline="") as file:
    write=csv.writer(file)
    for row in list:
        write.writerow(row)
```
