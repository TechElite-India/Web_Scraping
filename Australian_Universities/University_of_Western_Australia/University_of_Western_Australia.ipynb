```python
import requests
import bs4
import csv

list=[]
lis_h=['Course_Name','Course_Level','Course_Type','Cource_Description','University_Name','Link']
list.append(lis_h)
```

```python
res1=requests.get('http://handbooks.uwa.edu.au/search?type=courses') 
soup1=bs4.BeautifulSoup(res1.text,'lxml')
```

```python
s1=soup1.select('.filter-item')
for i in range(len(s1)):
    list.append([s1[i].a.h4.text,'','','','University of Western Australia',s1[i].a.get('href')])
    
list
```

```python
len(list)
```

```python
with open('University_of_Western_Australia.csv','w',newline="") as file:
    write=csv.writer(file)
    for row in list:
        write.writerow(row)
```
