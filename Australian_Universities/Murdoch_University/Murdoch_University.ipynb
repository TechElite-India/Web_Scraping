```python
import requests
import bs4
import csv

list=[]
lis_h=['Course_Name','Course_Level','Course_Type','Cource_Description','University_Name','Link']
list.append(lis_h)

res1=requests.get('https://search.murdoch.edu.au/s/search.html?collection=mu-course-search')
soup1=bs4.BeautifulSoup(res1.text,'lxml')
s1=soup1.select('h4')
s11=soup1.select('.card-text')
for i in range(len(s1)):
    list.append([s11[i].div.text,'',str(s1[i].text)[14:-12],'','Murdoch_University',s1[i].a.get('title')])
    
for j in range(1,26):
    res1=requests.get('https://search.murdoch.edu.au/s/search.html?collection=mu-course-search&start_rank='+str(j)+str(1))
    soup1=bs4.BeautifulSoup(res1.text,'lxml')
    s1=soup1.select('h4')
    s11=soup1.select('.card-text')
    for i in range(len(s1)):
        list.append([s11[i].div.text,'',str(s1[i].text)[14:-12],'','Murdoch_University',s1[i].a.get('title')])
        
list
```

```python
with open('Murdoch_University.csv','w',newline="") as file:
    write=csv.writer(file)
    for row in list:
        write.writerow(row)
```
