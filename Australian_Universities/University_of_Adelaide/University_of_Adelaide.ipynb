```python
import requests
import bs4
import csv

list=[]
lis_h=['Course_Name','Course_Level','Course_Type','Cource_Description','University_Name','Link']
list.append(lis_h)

res=requests.get('https://www.adelaide.edu.au/degree-finder/#ugrad-brse-tab')
soup=bs4.BeautifulSoup(res.text,'lxml')
s=soup.select('.c-media-object__column')
for j in range(14):
    if(j!=5 and j!=11):
        if(j<=4):
            cl='Undergraduate'
        elif(j<=10):
            cl='Postgraduate Coursework'
        else:
            cl='Postgraduate Research'
        s1=s[j].select('a')
        for k in range(1,len(s1)):
            res2=requests.get('https://www.adelaide.edu.au'+s1[k].get('href'))
            soup2=bs4.BeautifulSoup(res2.text,'lxml')
            s2=soup2.select('table')
            for l in range(len(s2)):
                s3=s2[l].select('a')
                for m in range(len(s3)):
                    list.append([s3[m].text,cl,s1[k].text,'','University of Adelaide','https://www.adelaide.edu.au'+s3[m].get('href')])
                    
list
```

```python
len(list)
```

```python
with open('University-of_Adelaide.csv','w',newline="") as file:
    write=csv.writer(file)
    for row in list:
        write.writerow(row)
```
