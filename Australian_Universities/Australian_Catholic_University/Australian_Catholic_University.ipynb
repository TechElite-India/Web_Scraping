```python
import requests
import bs4
import csv

list=[]
lis_h=['Course_Name','Course_Level','Course_Type','Cource_Description','University_Name','Link']
list.append(lis_h)

res1=requests.get('https://www.acu.edu.au/courses/courses')
soup1=bs4.BeautifulSoup(res1.text,'lxml')
s1=soup1.select('.study-area-odd')
s11=soup1.select('.study-area-even')
s=s1+s11
for j in s:
    res2=requests.get(j.a.get('href'))
    soup2=bs4.BeautifulSoup(res2.text,'lxml')
    s2=soup2.select('.course-list')
    for k in s2:
        list.append([str(k.a.text)[5:-8],'',j.a.text,'','Australian Catholic University',k.a.get('href')])
        
list
```

```python
len(list)
```

```python
with open('Australian_Catholic_University.csv','w',newline="") as file:
    write=csv.writer(file)
    for row in list:
        write.writerow(row)
```
