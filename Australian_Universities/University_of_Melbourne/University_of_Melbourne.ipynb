```python
import requests
import bs4
import csv

list=[]
lis_h=['Course_Name','Course_Level','Course_Type','Cource_Description','University_Name','Link']
list.append(lis_h)

res3=requests.get('https://study.unimelb.edu.au/find/')
soup3=bs4.BeautifulSoup(res3.text,'lxml')
for m in range(12):
    s3=soup3.select('#featured-interest-'+str(m))
    res2=requests.get('https://study.unimelb.edu.au'+s3[0].a.get('href'))
    soup2=bs4.BeautifulSoup(res2.text,'lxml')
    s2=soup2.select('.interest-container__related-list-item')
    for l in range(len(s2)):
        if(s2[l].a.get('class')[1]=='interest-item--tight'):
            res1=requests.get('https://study.unimelb.edu.au'+s2[l].a.get('href'))
            soup1=bs4.BeautifulSoup(res1.text,'lxml')
            s1=soup1.select('ul')
            for j in range(len(s1)):
                if(s1[j].get('class')[0]=='grid'):
                    for k in s1[j].select('a'):
                        list.append([k.select('span')[1].text,'',s3[0].span.text,'','University of Melbourne','https://study.unimelb.edu.au'+k.get('href')])
                    break
                    
list
```

```python
with open('University_of_Melbourne.csv','w',newline="") as file:
    write=csv.writer(file)
    for row in list:
        write.writerow(row)
```
