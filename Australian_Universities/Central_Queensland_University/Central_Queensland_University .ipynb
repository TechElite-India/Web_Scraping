```python
import requests
import bs4
import csv

res1=requests.get('https://www.cqu.edu.au/courses')
soup=bs4.BeautifulSoup(res1.text,'lxml')

s=soup.select('.area-blocks__link')

list=[]
lis_h=['Course_Name','Course_Level','Course_Type','University_Name','Link']
list.append(lis_h)

for j in s:
    res1=requests.get(j.get('href'))
    soup=bs4.BeautifulSoup(res1.text,'lxml')
    
    s1=soup.select('.box-featured__heading-link')
    s2=soup.select('.box-featured__level')
    s3=soup.select('.box-featured__description')
    
    
    for i in range(len(s1)-1):
        lis=[s1[i].text[7:(len(s1[i].text)-6)],s2[i].text,j.text,s3[i].text[1:(len(s3[i].text)-3)],'Bond University',j.get('href')]
        list.append(lis)

with open('Bond_University.csv','w') as file:
    write=csv.writer(file)
    for row in list:
        write.writerow(row)
```

```python
import requests
import bs4
import csv

res1=requests.get('https://www.cqu.edu.au/courses')
soup=bs4.BeautifulSoup(res1.text,'lxml')

s=soup.select('a')
list1=[]
list2=[]

for j in s[104:116]:
    list1.append(j.get('href'))
    list2.append(j.text)


list=[]
lis_h=['Course_Name','Course_Level','Course_Type','Cource_Description','University_Name','Link']
list.append(lis_h)

for j in range(len(list1)):
    res1=requests.get(list1[j])
    soup=bs4.BeautifulSoup(res1.text,'lxml')

    s1=soup.select('.sh-target')
    for z in range(len(s1)):
        for k in s1[z].select('a'):
            list.append([k.text,'',list2[j],'','Central Queensland University',k.get('href')])
list
```

```python
with open('Central_Queensland_University.csv','w',newline="") as file:
    write=csv.writer(file)
    for row in list:
        write.writerow(row)
```
