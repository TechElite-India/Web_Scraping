```python
import requests
import bs4
import csv

res1=requests.get('https://bond.edu.au/intl/future-students/study-bond/search-program')
soup=bs4.BeautifulSoup(res1.text,'lxml')

s=soup.select('.area-blocks__link')

list=[]
lis_h=['Course_Name','Course_Level','Course_Type','University_Name','Link']
list.append(lis_h)

for j in s:
    res1=requests.get(j.get('href'))
    soup=bs4.BeautifulSoup(res1.text,'lxml')
    
    s1=soup.select('.box-featured__heading-link')
    s2=soup.select('.box-featured__level')
    s3=soup.select('.box-featured__description')
    
    
    for i in range(len(s1)-1):
        lis=[s1[i].text[7:(len(s1[i].text)-6)],s2[i].text,j.text,s3[i].text[1:(len(s3[i].text)-3)],'Bond University',j.get('href')]
        list.append(lis)

with open('Bond_University.csv','w') as file:
    write=csv.writer(file)
    for row in list:
        write.writerow(row)
```

```python
import requests
import bs4
import csv

res1=requests.get('https://bond.edu.au/intl/future-students/study-bond/search-program')
soup=bs4.BeautifulSoup(res1.text,'lxml')

list=[]
s=soup.select('a')

for j in s:
    list.append('https://bond.edu.au' + str(j.get('href')))
    
list1=list[37:89]

list_f=[]
lis_h=['Course_Name','Course_Level','Course_Type','Cource_Description','University_Name','Link']
list_f.append(lis_h)

for link in list1:
    res1=requests.get(link)
    soup=bs4.BeautifulSoup(res1.text,'lxml')

    s=soup.select('body')
    cn=""
    cl=""
    ct=""

    s=str(s)

    for i in s[207:]:
        if(i=='"'):
            break
        cn+=i

    for i in s[(207+len(cn)+21):]:
        if(i=='"'):
            break
        cl+=i
    
    for i in s[(207+len(cn)+21+len(cl)+19):]:
        if(i=='"'):
            break
        ct+=i
    
    if(ct!='nk'):
        list_f.append([cn,cl,ct,'','Bond University',link])
        
        
list2=list[98:189]      
for link in list2:    
    res1=requests.get(link)
    soup=bs4.BeautifulSoup(res1.text,'lxml')

    s=soup.select('body')
    cn=""
    cl=""
    ct=""

    s=str(s)

    for i in s[206:]:
        if(i=='"'):
            break
        cn+=i

    for i in s[(206+len(cn)+21):]:
        if(i=='"'):
            break
        cl+=i

    for i in s[(206+len(cn)+21+len(cl)+19):]:
        if(i=='"'):
            break
        ct+=i

    if(ct!='nk'):
        list_f.append([cn,cl,ct,'','Bond University',link])
    
with open('Bond_University.csv','w',newline="") as file:
    write=csv.writer(file)
    for row in list_f:
        write.writerow(row)
```
