```python
import requests
import bs4
import csv

list=[]
lis_h=['Course_Name','Course_Level','Course_Type','Cource_Description','University_Name','Link']
list.append(lis_h)

res1=requests.get('https://www.jcu.edu.au/courses-and-study')
soup1=bs4.BeautifulSoup(res1.text,'lxml')
s1=soup1.select('.megamenu__level-3-item')
for j in range(26,31):
    res2=requests.get(s1[j].a.get('href'))
    soup2=bs4.BeautifulSoup(res2.text,'lxml')
    s2=soup2.select('.course__title')
    for k in range(len(s2)):
        res3=requests.get(s2[k].a.get('href'))
        soup3=bs4.BeautifulSoup(res3.text,'lxml')
        s3=soup3.select('.coursedetails__intro')
        if(len(s3)!=0):
            list.append([s2[k].a.text,str(s1[j].a.text)[85:-80],'',s3[0].text,'James Cook University',s2[k].a.get('href')])
        else:
            list.append([s2[k].a.text,str(s1[j].a.text)[85:-80],'','','James Cook University',s2[k].a.get('href')])
list
```

```python
with open('James_Cook_University.csv','w',encoding="utf-8",newline="") as file:
    write=csv.writer(file)
    for row in list:
        write.writerow(row)
```
