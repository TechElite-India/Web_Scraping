```python
import requests
import bs4
import csv

list=[]
lis_h=['Course_Name','Course_Level','Course_Type','Cource_Description','University_Name','Link']
list.append(lis_h)

list1=[]

res1=requests.get('https://futurestudents.csu.edu.au/courses/all')
soup=bs4.BeautifulSoup(res1.text,'lxml')

s=soup.select('h3')
for i in s[1:]:
    list1.append(i.text)
    

list2=[]
list3=[]
s=soup.select('a')
for i in s[68:442]:
    list2.append(i.get('href'))
    list3.append(i.text)
    

x=0
z=1
m=0
n=0
for k in range(len(list2)):
    if(len(list3[k])!=0):
        if(list3[k][0]=='A' or list3[k][0]=='B'):
            if(z==2):
                x=x+1
                if(x==8 and m==0):
                    x=7
                    m=1
                if(x==14 and n==0):
                    x=13
                    n=1
                z=1
        else:
            z=2
        list.append([list3[k],'',list1[x],'','Charles Sturt University',list2[k]])
    
with open('Charles_Sturt_University.csv','w',newline="") as file:
    write=csv.writer(file)
    for row in list:
        write.writerow(row)
```
