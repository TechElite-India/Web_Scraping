```python
import requests
import bs4
import csv

list=[]
lis_h=['Course_Name','Course_Level','Course_Type','Cource_Description','University_Name','Link']
list.append(lis_h)

res1=requests.get('https://www.vu.edu.au/study-at-vu/courses/browse-study-areas/all-courses-a-to-z')
soup1=bs4.BeautifulSoup(res1.text,'lxml')
s1=soup1.select('.course-list-wrapper')
for i in range(len(s1)):    
    for j in s1[i].select('a'):
        list.append([j.text,'','','','Victoria University','https://www.vu.edu.au'+j.get('href')])
        
list
```

```python
len(list)
```

```python
with open('Victoria_University.csv','w',newline="") as file:
    write=csv.writer(file)
    for row in list:
        write.writerow(row)
```
