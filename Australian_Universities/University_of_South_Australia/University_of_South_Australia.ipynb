```python
import requests
import bs4
import csv

list=[]
lis_h=['Course_Name','Course_Level','Course_Type','Cource_Description','University_Name','Link']
list.append(lis_h)

res=requests.get('http://study.unisa.edu.au/?_ga=2.151726393.1465123207.1540366871-2127559637.1540366871#undergraduate')
soup=bs4.BeautifulSoup(res.text,'lxml')
s=soup.select('a')
for j in s[34:74]:
    res1=requests.get('http://study.unisa.edu.au'+j.get('href'))
    soup1=bs4.BeautifulSoup(res1.text,'lxml')
    s1=soup1.select('table')
    for i in range(len(s1)):
        k=1
        while True:
            if(k<len(s1[i].select('td'))):
                list.append([str(s1[i].select('td')[k].text)[1:-1],'',str(j.text)[7:-1],'','University of South Australia','http://study.unisa.edu.au'+s1[i].select('td')[k].a.get('href')])
            else:
                break
            k+=5
            
list
```

```python
len(list)
```

```python
with open('University_of_South_Australia.csv','w',newline="") as file:
    write=csv.writer(file)
    for row in list:
        write.writerow(row)
```
