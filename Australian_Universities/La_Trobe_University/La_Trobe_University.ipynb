```python
import requests
import bs4
import csv

list=[]
lis_h=['Course_Name','Course_Level','Course_Type','Cource_Description','University_Name','Link']
list.append(lis_h)

res2=requests.get('https://www.latrobe.edu.au/courses/a-z')
soup2=bs4.BeautifulSoup(res2.text,'lxml')

s2=soup2.select('.beige-arrows')
for j in s2[0].select('a'):
    res1=requests.get(j.get('href'))
    soup1=bs4.BeautifulSoup(res1.text,'lxml')
    
    s1=soup1.select('.course-url')
    for i in s1:
        list.append([i.text,'',j.text,'','La Trobe University',i.get('href')])
        
list
```

```python
len(list)
```

```python
with open('La_Trobe_University.csv','w',newline="") as file:
    write=csv.writer(file)
    for row in list:
        write.writerow(row)
```
