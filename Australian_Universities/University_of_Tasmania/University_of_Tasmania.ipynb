```python
import requests
import bs4
import csv

list=[]
lis_h=['Course_Name','Course_Level','Course_Type','Cource_Description','University_Name','Link']
list.append(lis_h)
```

```python
lk=['undergraduate','postgraduate']
for z in range(2):
    res1=requests.get('http://www.utas.edu.au/courses/'+lk[z]) 
    soup1=bs4.BeautifulSoup(res1.text,'lxml')
    s1=soup1.select('.course-list')
    for i in s1[0].select('a'):
        if(i.get('href')!='#'):
            str=''
            for j in i.text:
                if(j=='\n'):
                    break
                str+=j
            list.append([str,lk[z][0].upper()+lk[z][1:],'','','University of Tasmania',i.get('href')])
list
```

```python
len(list)
```

```python
with open('University_of_Tasmania.csv','w',newline="") as file:
    write=csv.writer(file)
    for row in list:
        write.writerow(row)
```
