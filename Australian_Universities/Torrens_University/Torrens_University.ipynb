```python
import requests
import bs4
import csv

list=[]
lis_h=['Course_Name','Course_Level','Course_Type','Cource_Description','University_Name','Link']
list.append(lis_h)

s2=soup2.select('.sub-menu')
for k in s2[0].select('a'):
    res1=requests.get(k.get('href')) 
    soup1=bs4.BeautifulSoup(res1.text,'lxml')
    
    link=[]
    s1=soup1.select('.course-wrap')
    for j in s1[0].select('a'):
        if(j.text=='View Course Details'):
            link.append(j.get('href'))
   
    name=[]
    s11=soup1.select('h3')
    for i in s11:
        if(i.text!='Undergraduate' and i.text!='Postgraduate' and i.text!='Short Courses'):
            name.append(i.text)
    for z in range(len(link)):
        list.append([name[z],'',k.text,'','Torrens University',link[z]])
    if(len(link)==len(name)):
        print(0)
    else:
        print(k.text)
        
list
```

```python
with open('Torrens_University.csv','w',newline="") as file:
    write=csv.writer(file)
    for row in list:
        write.writerow(row)
```
