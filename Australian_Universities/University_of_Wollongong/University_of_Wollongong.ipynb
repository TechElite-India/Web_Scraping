```python
import requests
import bs4
import csv

list=[]
lis_h=['Course_Name','Course_Level','Course_Type','Cource_Description','University_Name','Link']
list.append(lis_h)

s1=soup1.select('.study-options')
for i in s1[0].select('a'):
    res2=requests.get('https://coursefinder.uow.edu.au'+i.get('href'))
    soup2=bs4.BeautifulSoup(res2.text,'lxml')
    s2=soup2.select('.two-column')
    for j in range(len(s2)):
        for k in s2[j].select('a'):
            list.append([k.text,'',str(i.text)[1:],'','University of Wollongong',k.get('href')])
            
list
```

```python
with open('University_of_Wollongong.csv','w',newline="") as file:
    write=csv.writer(file)
    for row in list:
        write.writerow(row)
```
