```python
import requests
import bs4
import csv

list=[]
lis_h=['Course_Name','Course_Level','Course_Type','Cource_Description','University_Name','Link']
list.append(lis_h)

res=requests.get('https://www.qut.edu.au')
soup=bs4.BeautifulSoup(res.text,'lxml')
s=soup.select('.arrow-links')
for i in s[0].select('a'):
    res2=requests.get(i.get('href'))
    soup2=bs4.BeautifulSoup(res2.text,'lxml')
    s2=soup2.select('.course-list')
    for k in range(len(s2)):
        if(s2[k].text!='\n'):
            lt=[]
            s3=s2[k].select('p')
            for m in range(len(s3)):
                if(m%2!=0):
                    lt.append(s3[m].text)
                    
            s4=s2[k].select('a')
            for l in range(len(s4)):
                if(len(lt)==len(s4)):
                    list.append([s4[l].text,'',i.text,lt[l],'Queensland University of Technology',s4[l].get('href')])
                else:
                    list.append([s4[l].text,'',i.text,'','Queensland University of Technology',s4[l].get('href')])
        
list
```

```python
len(list)
```

```python
with open('Queensland_University_of_Technology.csv','w',encoding="utf-8",newline="") as file:
    write=csv.writer(file)
    for row in list:
        write.writerow(row)
```
