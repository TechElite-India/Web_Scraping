```python
import requests
import bs4
import csv

list=[]
lis_h=['Course_Name','Course_Level','Course_Type','Cource_Description','University_Name','Link']
list.append(lis_h)
```

```python
res=requests.get('https://my.une.edu.au/courses/2019/courses/browse')
soup=bs4.BeautifulSoup(res.text,'lxml')
```

```python
s=soup.select('.nth-child-even ')
for i in s[2:]:
    print(str(i.a.text)[2:-2])
    print('https://my.une.edu.au'+i.a.get('href'))
```

```python
for i in s[2:]:
    list.append([str(i.a.text)[2:-2],'','','','University of New England','https://my.une.edu.au'+i.a.get('href')])
```

```python
len(list)
```

```python
s=soup.select('.nth-child-odd ')
for i in s[2:]:
    print(str(i.a.text)[2:-2])
    print('https://my.une.edu.au'+i.a.get('href'))
```

```python
for i in s[2:]:
    list.append([str(i.a.text)[2:-2],'','','','University of New England','https://my.une.edu.au'+i.a.get('href')])
```

```python
len(list)
```

```python
with open('University_of_New_England.csv','w',newline="") as file:
    write=csv.writer(file)
    for row in list:
        write.writerow(row)
```
