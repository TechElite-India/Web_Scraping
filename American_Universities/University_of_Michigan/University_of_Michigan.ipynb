```python
lis=['Course','URL','University']
info=[]
info.append(lis)

from selenium import webdriver
import csv

chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--no-sandbox')
driver = webdriver.Chrome("C:\\Users\MOHAN KUMAR SAH\Documents\My Work\PakkaIndia\chromedriver",chrome_options=chrome_options)

driver.get('https://www.si.umich.edu/programs/courses/catalog')

data=driver.find_elements_by_tag_name('tbody')
data1=data[2].find_elements_by_tag_name('a')
for i in range(len(data1)):
    if(i%2!=0):
        c=data1[i].text
        url=data1[i].get_attribute('href')
        uni='University_of_Michigan'
        
        info.append([c,url,uni])
        print(c,url,uni)
        
info
```

```python
len(info)
```

```python
with open('University_of_Michigan.csv','w',encoding="utf-8",newline="") as file:
    write=csv.writer(file)
    for row in info:
        write.writerow(row)
```
