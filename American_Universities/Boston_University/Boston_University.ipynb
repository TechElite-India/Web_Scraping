```python
from selenium import webdriver
import csv

chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--no-sandbox')
driver = webdriver.Chrome("C:\\Users\MOHAN KUMAR SAH\Documents\My Work\PakkaIndia\chromedriver",chrome_options=chrome_options)

lis=['Course','URL','University']
info=[]
info.append(lis)

for i in range(103):
    driver.get('https://www.bu.edu/academics/cas/courses/'+str(i)+'/')
    data1=driver.find_elements_by_class_name('course-feed')
    data2=data1[0].find_elements_by_tag_name('a')
    for j in range(len(data2)):
        c=data2[j].text
        url=data2[j].get_attribute('href')
        if(c!='BU Hub' and c!='Learn More'):
            info.append([c,url,'Boston_University'])
            print(c,url)
```

```python
len(info)
```

```python
with open('Boston_University.csv','w',encoding="utf-8",newline="") as file:
    write=csv.writer(file)
    for row in info:
        write.writerow(row)
```
