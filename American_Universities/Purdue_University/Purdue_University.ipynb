```python
lis=['Course','URL','University']
info=[]
info.append(lis)

from selenium import webdriver
import csv

chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--no-sandbox')
driver = webdriver.Chrome("C:\\Users\MOHAN KUMAR SAH\Documents\My Work\PakkaIndia\chromedriver",chrome_options=chrome_options)

driver.get('http://catalog.purdue.edu/content.php?catoid=9&navoid=11239')

data=driver.find_elements_by_tag_name('ul')
for i in range(6,66):
    data1=data[i].find_elements_by_tag_name('a')
    for j in range(len(data1)):
        url=data1[j].get_attribute('href')
        c=data1[j].text
        
        info.append([c,url,'Purdue_University'])
        print(c,url)
        
driver.quit()
```

```python
len(info)
```

```python
with open('Purdue_University.csv','w',encoding="utf-8",newline="") as file:
    write=csv.writer(file)
    for row in info:
        write.writerow(row)
```
