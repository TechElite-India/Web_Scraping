```python
from selenium import webdriver
import csv
import time

driver = webdriver.Chrome("C:\\Users\MOHAN KUMAR SAH\Documents\My Work\PakkaIndia\chromedriver")
driver.get('https://www.4icu.org/us/')

lis=['Rank','University','URL']
info=[]
info.append(lis)

data=driver.find_elements_by_tag_name("tbody")
data1=data[0].find_elements_by_tag_name("tr")
l=len(data1)
for i in range(l-1):
    time.sleep(1)
    
    data=driver.find_elements_by_tag_name("tbody")
    data1=data[0].find_elements_by_tag_name("tr")
    
    data2=data1[i].find_elements_by_tag_name("td")
    
    rank=data2[0].text
    uni=data2[1].text
    
    link=data2[1].find_element_by_tag_name('a')
    
    #driver.get(link)        # wrong
    link.click()
    data3=driver.find_elements_by_link_text(uni)
    if(len(data3)==3):
        url=data3[1].get_attribute('href')
    else:
        print(rank,'Problem')
        url=''
    driver.back()
    
    info.append([rank,uni,url])
    print(rank,uni,url)
    
driver.quit()
```

```python
with open('2018_US_University_List.csv','w',encoding="utf-8",newline="") as file:
    write=csv.writer(file)
    for row in info:
        write.writerow(row)
```
